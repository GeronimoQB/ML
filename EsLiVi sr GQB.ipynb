```r
# codigo limpieza y manipulación 
casos_diarios <-  read.csv(file = "Casos_Diarios_Estado_Nacional.csv")
casos_diario_list <- gather(casos_diarios,key ="dias",value="infectados",4:856)

normalizarfecha <- function(fecha){
return(str_replace_all(fecha,"X","")  %>% str_replace_all("\\.","/"))
}

cod_Meses <- function(x) {

labels_meses <- c(
    "Ene","Feb","Mar","Abr","May","Jun",
    "Jul","Ago","Sep","Oct","Nov","Dic")

    return(factor(x,levels = (1:12),labels = labels_meses))
    
    }


cod_dias <- function(x) {
    
labels_dias <- c(
    "Miercoles","Jueves","Viernes","Sabado","Domingo","Lunes",
    "Martes")

    return(factor(x,levels = (1:7), labels = labels_dias))
    
    }



fecha = map(casos_diario_list[4],normalizarfecha)
casos_diarios_fecha <- cbind(casos_diario_list[c(TRUE,TRUE,TRUE,FALSE,TRUE)],fecha)
casos_diarios_fecha <- casos_diarios_fecha[c(1,2,3,5,4)]

datos <- separate( data = casos_diarios_fecha,
                   col =dias,
                   into = c("dia","mes","año"),
                   sep = '/',
                   remove = FALSE)



#recordatorio
datos <- datos %>% mutate(mest_txt = cod_Meses(as.integer(mes))) %>% mutate(consecutivo = 1:28149) %>% transform(dias = as.Date(dias,"%d/%m/%Y"))
datosnacional <- datos %>% filter(nombre == "Nacional") %>% mutate(consecutivo = 1:853)

#datosnacional

#datos <- datos %>% mutate(diasem = cod_dias(as.integer(mes))) %>% mutate(consecutivo = 1:28149) %>% transform(dias = as.Date(dias,"%d/%m/%Y"))

#datos$diasem <- as.character(ifelse(datos$diasem == "Sabado" | datos$diasem == "Domingo",  "NA", datos$diasem))

#datos <- factor(datos, levels = c("Monday", "Tuesday", "Wednesday", 
                          #"Thursday", "Friday", "Saturday", "Sunday"), ordered = TRUE)


#datos

consecutivo <- 1:853
dia_sem <- 1:7

data_com <- cbind(consecutivo, dia_sem)

salida2 <- merge(datosnacional, as.data.frame(data_com), by = c("consecutivo"))

salida2 <- salida2 %>% mutate(dias_txt = cod_dias(as.integer(dia_sem)))
salida2$infectados[salida2$dias_txt == "Sabado"] <- NA
salida2$infectados[salida2$dias_txt == "Domingo"] <- NA

#salida2$infectados[(salida2$dia_sem %% 4) == 0] <- NA
#salida2$infectados[(salida2$dia_sem %% 5) == 0] <- NA

salida2

#datos <- datos %>%
 #   mutate()
#str(datos)
#unique(datos$mes)
```
