```r

datos <- cbind(cluster = resultado$cluster, casos_diarios)
casos_diartios_list <- gather(datos, key = "dias", value = "infectados", 5:857)

#head(casos_diartios_list)

normalizarfecha <- function(fecha){
return(str_replace_all(fecha,"X","")  %>% str_replace_all("\\.","/"))
}

cod_Meses <- function(x) {

labels_meses <- c(
    "Ene","Feb","Mar","Abr","May","Jun",
    "Jul","Ago","Sep","Oct","Nov","Dic")

    return(factor(x,levels = (1:12),labels = labels_meses))
    
    }

fecha = map(casos_diartios_list[5],normalizarfecha)
#str(casos_diartios_list)
casos_diarios_fecha <- cbind(casos_diartios_list[c(TRUE, TRUE, TRUE, TRUE, FALSE, TRUE)],fecha)

#casos_diarios_fecha

datos <- separate (data = casos_diarios_fecha,
                  col = dias,
                  into = c("dia", "mes", "años"),
                  sep = '/',
                  remove = FALSE)

#recordatorio
datos <- datos %>% mutate(mest_txt = cod_Meses(as.integer(mes))) %>% mutate(consecutivo = 1:28149) %>% filter(nombre != "Nacional") %>%
filter (cluster %in% c(6, 2))

datos %>% ggplot(aes(x = as.Date(dias, "%d/%m/%Y"), y = infectados, group = cluster, color = as.factor(cluster))) +
        geom_point()


```
