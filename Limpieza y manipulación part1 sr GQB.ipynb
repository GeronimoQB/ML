### purr

usando la función map obten los años bisiestos del siglo XX

```r
is_leap <- function(year){
    return((((year %% 4) == 0) && ((year %% 100)!=0)) || ((year %% 400)== 0))
}

#map

sXX <- 1901:2000
mask <- map(sXX, is_leap)

leaps_sXX <- sXX[unlist(mask)]
leaps_sXX
```

```r
casos_diarios <-  read.csv(file = "Casos_Diarios_Estado_Nacional.csv")
#str(casos_diarios)
#length(names(casos_diarios)) #856 columnas
casos_diarios_list <- gather(casos_diarios, key = "dias", value = "infectados", 4:856)
casos_diarios_list

#####
normalizarfecha <- function(fecha)
    {
    return(str_replace_all(fecha, "X", "") %>% str_replace_all("\\.", "/"))
    }

#a <- normalizarfecha(casos_diarios_list[4])
#a

#a <- casos_diarios_list[4] %>% str_replace_all("X", "") %>% str_replace_all("\\.", "/")
#a
#vect <- str_replace_all(casos_diarios_list[4], "X", "")
#str_replace_all(vect, "\\.", "/")

fecha = map(casos_diarios_list[4], normalizarfecha)
#fecha

casos_diarios_fecha <- cbind(casos_diarios_list[, c(TRUE, TRUE, TRUE, FALSE, TRUE )], fecha)
casos_diarios_fecha <- casos_diarios_fecha[, c(1, 2, 3, 4, 5)]
#casos_diarios_fecha

salida <- separate(data = casos_diarios_fecha,
                    col = dias,
                    into = c("dia", "mes", "año"),
                   sep = "/")
salida
```
