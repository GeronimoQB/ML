### obtener las primeras filas

```r
head(mpg)
```

### unir data frames


```r
mpg1 <- select(mpg, name, mpg, origin, model_year, weight, cylinders)
mpg2 <- select(mpg, name, mpg, model_year, displacement, horsepower)
merge(mpg1,mpg2, by = c("name", "mpg", "model_year"))
```

### encontrar missing values



```r
mpg %>% filter(is.na(weight))
```

### Eliminar tuplas con valores NA

```r
str(mpg)
mpg_drop <- mpg %>% drop_na(mpg)
str(mpg_drop)
```

### tidyr fill
Esta función completa los valores NA con los valores adyacentes, se puede especificar la busqueda hacia arriba o abajo de la fila con NA. Si no se especifica una dirección se toma “down” 

```r
mpg_filled <- mpg %>% fill(acceleration)
summary(mpg_filled)
```

### replace NA

```r
mpg %>% mutate(weight2 = replace_na(weight, 30000))
```

```r
mpg %>% mutate(acceleration2 = replace_na(acceleration, 30000))
```
